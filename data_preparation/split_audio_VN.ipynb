```python
import os
import shutil
import pandas as pd

# --- CẤU HÌNH ĐƯỜNG DẪN (Bạn điền vào đây) ---
# Đường dẫn đến file metadata .csv
metadata_path = r'D:\Study\7-SP26\DATxSLP\metadata_other_languages.csv'

# Đường dẫn folder gốc chứa file .wav
source_folder = r'D:\Study\7-SP26\DATxSLP\Data_after_cut\test_output'

# Đường dẫn folder mới để lưu các file copy
destination_folder = r'D:\Study\7-SP26\DATxSLP\VN'
# ---------------------------------------------

def filter_and_copy_files():
    # 1. Tạo folder đích nếu chưa tồn tại
    if not os.path.exists(destination_folder):
        os.makedirs(destination_folder)
        print(f"Đã tạo folder mới: {destination_folder}")

    # 2. Đọc file metadata và lấy danh sách tên file cần bỏ qua
    try:
        df = pd.read_csv(metadata_path)
        # Chuyển cột file_name thành set để tra cứu cho nhanh
        # Lưu ý: Đảm bảo tên trong csv và tên file thực tế giống nhau (cùng có đuôi .wav hoặc cùng không)
        files_to_skip = set(df['file_name'].astype(str).str.strip())
        print(f"Tìm thấy {len(files_to_skip)} file trong metadata cần bỏ qua.")
    except Exception as e:
        print(f"Lỗi khi đọc file CSV: {e}")
        return

    # 3. Duyệt qua folder gốc và copy
    copied_count = 0
    skipped_count = 0
    
    # Lấy danh sách tất cả file trong folder gốc
    files_in_source = os.listdir(source_folder)

    print("Đang xử lý...")
    for filename in files_in_source:
        # Chỉ xử lý file .wav (hoặc bạn có thể bỏ dòng này nếu muốn copy mọi loại file)
        if filename.lower().endswith('.wav'):
            
            # KIỂM TRA LOGIC:
            # Nếu file nằm trong danh sách metadata (files_to_skip) -> Bỏ qua
            # Nếu KHÔNG nằm trong danh sách -> Copy
            if filename in files_to_skip:
                skipped_count += 1
            else:
                src_path = os.path.join(source_folder, filename)
                dst_path = os.path.join(destination_folder, filename)
                
                shutil.copy2(src_path, dst_path) # copy2 giữ nguyên metadata của file (ngày tạo, v.v.)
                copied_count += 1

    # 4. Thông báo kết quả
    print("-" * 30)
    print("HOÀN TẤT!")
    print(f"Số file đã bỏ qua (có trong metadata): {skipped_count}")
    print(f"Số file đã copy sang folder mới: {copied_count}")

if __name__ == "__main__":
    filter_and_copy_files()
```
