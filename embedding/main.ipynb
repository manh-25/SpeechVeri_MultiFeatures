### Ki·ªÉm tra th·ª≠ d·ªØ li·ªáu sau khi l∆∞u

```python
import torch
import os
import glob

MODELS = ["wavlm", "hubert", "wav2vec2"]
BASE_SAVE_DIR = "./embeddings_shards"

print(f"üìä B√ÅO C√ÅO T·ªîNG H·ª¢P TR√çCH XU·∫§T (Dataset: 258,168 files)")
print("="*60)

for m in MODELS:
    model_dir = os.path.join(BASE_SAVE_DIR, m)
    shard_files = glob.glob(os.path.join(model_dir, f"{m}_shard_*.pt"))
    
    if not shard_files:
        print(f"‚ùå Model {m:10}: Kh√¥ng t√¨m th·∫•y d·ªØ li·ªáu.")
        continue
        
    total_samples = 0
    # Ch·ªâ load metadata t·ª´ shard ƒë·∫ßu ti√™n ƒë·ªÉ xem c·∫•u tr√∫c
    sample_data = torch.load(shard_files[0], map_location='cpu')
    
    for f in shard_files:
        # D√πng k·ªπ thu·∫≠t load nh·∫π ƒë·ªÉ ƒë·∫øm s·ªë l∆∞·ª£ng
        data = torch.load(f, map_location='cpu')
        total_samples += len(data['filenames'])
        del data
        
    status = "‚úÖ ƒê·ª¶" if total_samples >= 258168 else "‚ö†Ô∏è THI·∫æU"
    
    emb_shape = sample_data['embeddings'].shape
    print(f"üîπ Model {m:10} | Shards: {len(shard_files):2} | T·ªïng m·∫´u: {total_samples:,} | {status}")
    print(f"   Structure: {emb_shape[1]} layers x {emb_shape[2]} dims | Dtype: {sample_data['embeddings'].dtype}")
    print("-"*60)
    
    del sample_data
```
